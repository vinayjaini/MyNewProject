```python
#importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Supress Warnings

import warnings
warnings.filterwarnings('ignore')
```

```python
loan = pd.read_csv('loan.csv')
loan
```

```python
#inspect data
loan.describe()
```

```python
#inspect data
loan.shape
```

```python
#Column data types
loan.dtypes
```

```python
#Cleaning data, inspect null values
loan.isnull().sum(axis=0)
```

```python
#listing columns which have 3000 null values
column_list = loan.columns[loan.isnull().sum(axis=0) > 3000].tolist()
column_list
```

```python
#Dropping columns which have null values 
loan.drop(axis=1, columns=column_list, inplace=True)

loan
```

```python
#Cleaning data, inspect null values
loan.isnull().sum(axis=0)
```

```python
#Removing droping rows which contain null values
loan.dropna(axis=0, how='any', inplace=True)
loan.isnull().sum(axis=0)
```

```python
#Calculating % of data retained after removing null values
round((100*(loan.shape[0]/39717)),2)
```

```python
loan
```

```python
#Extratcing months from term
term_n = loan['term'].str.split(' ', n=2, expand=True)
loan['term_n'] = term_n[1]
loan['term_n'] = loan['term_n'].astype('int64', inplace=True)
loan
```

```python
#extract int_rate by spliting on %
loan['int_rate'] = loan['int_rate'].str.split('%', n = 1, expand= True) 
loan['int_rate'] = loan['int_rate'].astype('float64', inplace=True)
loan
```

```python
#Fill null values in emp_title with XX and checking number of null values
loan['emp_title'].fillna('XX', inplace=True)
loan['emp_title'].isnull().sum()
```

```python
#Extracting employee work experinece from emp_length
emp_title_for_lenght_isna = loan[loan['emp_length'].isnull()]
np.unique(emp_title_for_lenght_isna['emp_title'])
```

```python
#Fill null values in emp_lenght with default value as 0
loan['emp_length'].isnull().sum()
loan['emp_length'].fillna('-1', inplace=True)
loan['emp_length'].isnull().sum()

```

```python
#Find unique values in emp_length 
#np.unique(loan['emp_length'].str)
loan.emp_length.unique()
```

```python
#Extract emp_length as integer value '< 1 year' as '-1', '10+ years' as 11 rest with years
loan.emp_length.replace(['10+ years','< 1 year','1 year','3 years','8 years','9 years','4 years','5 years','6 years','2 years','7 years'] ,['11','0','1','3','8','9','4','5','6','2','7'] , inplace = True)
loan.emp_length.unique()
```

```python
#Converting emp_length to int
loan['emp_length'] = loan['emp_length'].astype('int64', inplace=True)
loan.dtypes
```

```python
#Find if there are still null values exits
loan.isnull().sum(axis=0)
```

```python
#Fill null values of title 
loan['title'].fillna('XX', inplace=True)
```

```python
#removing 'XX' from zipcode and changing datatype to int
zipcodes = loan['zip_code'].str.split('xx', n=1, expand=True)
loan['zip_code'] = zipcodes[0]
loan['zip_code'] = loan['zip_code'].astype('int64', inplace=True)
```

```python
#revol_util there are only 50 null will replace them with avg rate
revol = loan['revol_util'].str.split('%',n=1,expand=True)
loan['revol_util'] = revol[0]
loan['revol_util'] = loan['revol_util'].astype('float64', inplace=True)
```

```python
#Check for null values
loan.isnull().sum(axis=0)
```

```python
#Fill revil_util null values with mean of values
loan['revol_util'].fillna((loan['revol_util'].mean()), inplace = True)
loan.isnull().sum(axis=0)
```
